```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
steps = 100

def brown(steps):
    origin = np.array([0,0,0])
    n = 0

    movement = np.zeros((steps+1,3))
    
    while n < steps:
        coord = origin
        direction = np.random.randint(0,3)

        step = 0
        while step == 0:
            step = np.random.randint(-1,2)
            
        coord[direction] = coord[direction] + step
        origin = coord
        n += 1

        movement[n] = origin
    length = np.sqrt(origin[0]**2 + origin[1]**2 + origin[2]**2) #dx prop to sqrt(t)

    return length, movement, n
```

```python
length, move, n = brown(steps)
```

```python
plt.plot(move[:,0],move[:,2])
```

```python
ax = plt.figure().add_subplot(projection='3d')

x = move[:,0]
y = move[:,1]
z = move[:,2]

ax.plot(x, y, z, label='parametric curve')
```

```python
def brown2d(steps):
    origin = np.array([0,0])
    n = 0

    movement = np.zeros((steps+1,2))
    
    while n < steps:
        coord = origin
        direction = np.random.randint(0,2)

        step = 0
        while step == 0:
            step = np.random.randint(-1,2)
            
        coord[direction] = coord[direction] + step
        origin = coord
        n += 1

        movement[n] = origin
    length = np.sqrt(origin[0]**2 + origin[1]**2)

    return length, movement
```

```python
l, m = brown2d(100)
```

```python
x = m[:,0]
y = m[:,1]

plt.plot(x,y)
```

```python
l = []
m = []
for i in range(5000):
    ll, m = brown2d(500)
    l.append(ll)
    
```

```python
plt.hist(l, bins = 75, histtype = 'step')
#plt.axvline(x = np.sqrt(500), c = 'red', label = 'Actual average:{}'.format(np.round(np.sqrt(500), 3)))
#plt.axvline(x = np.mean(l), c = 'blue', label = 'Measured average:{}'.format(np.round(np.mean(l), 3)))
plt.title('Brown 2d')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Counts')
plt.legend()
#plt.savefig('pics/2d_brown.png', dpi = 300)
```

```python
le = []
for i in range(5000):
    ll, mm = brown2d(500)
    le.append(ll)
le100 = []
for i in range(5000):
    ll, mm = brown2d(100)
    le100.append(ll)
    
le200 = []
for i in range(5000):
    ll, mm = brown2d(200)
    le200.append(ll)
    
le300 = []
for i in range(5000):
    ll, mm = brown2d(300)
    le300.append(ll)

le400 = []
for i in range(5000):
    ll, mm = brown2d(400)
    le400.append(ll)
```

```python
plt.hist(le, bins = 75, histtype = 'step');
plt.axvline(x = np.sqrt(500), c = 'red', label = 'Actual average:{}'.format(np.round(np.sqrt(500), 3)))
plt.axvline(x = np.mean(le), c = 'blue', label = 'Measured average:{}'.format(np.round(np.mean(le), 3)))
plt.title('Brown 3d')
plt.xlabel(r'$\Delta x$')
plt.ylabel('Counts')
plt.legend()
#plt.savefig('pics/3d_brown.png', dpi = 300)
```
